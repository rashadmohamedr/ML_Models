```python
file_path = '../input/students-score-dataset-linear-regression/student_scores.csv'
data = pd.read_csv(file_path)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(np.array(data.Hours).reshape(-1,1),
                                               np.array(data.Scores),
                                               random_state=42,
                                               test_size=0.2)
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
LRModel=LinearRegression()
LRModel.fit(X_train,y_train)
y_pred=LRModel.predict(X_test)
print(LRModel.score(X_test,y_test))
```

```python
output = pd.DataFrame({'Hours': X_test.flatten(),
                       'Scores': y_pred})
output.to_csv('submission.csv', index=False)
```
